```python
import pandas as pd

# Read the Excel file into a pandas DataFrame
data = pd.read_excel("pokemon_data.xlsx")

# Define a function to parse the 'spawn_time' values
def parse_spawn_time(time_str):
    if isinstance(time_str, str):
        parts = time_str.split(':')
        if len(parts) >= 2:
            hours = int(parts[0])
            minutes = int(parts[1])
            return pd.Timedelta(hours=hours, minutes=minutes)
    return pd.NaT

# Apply the parse_spawn_time function to the 'spawn_time' column
data['spawn_time'] = data['spawn_time'].apply(parse_spawn_time)
print(data['spawn_time'])
# Question 5: Get all Pokemons whose spawn time is less than 300 seconds
max_spawn_time = pd.Timedelta(seconds=300)
less_than_300s_spawn_time = data[data['spawn_time'] < max_spawn_time]
print("Pokemons with spawn time less than 300 seconds:")
print(less_than_300s_spawn_time[['name', 'spawn_time']])
```

```python
min_capabilities = 2
more_than_2_capabilities = data[data['type'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0) > min_capabilities]
print("Pokemons with more than two types of capabilities:")
print(more_than_2_capabilities[['name', 'type']])

```
